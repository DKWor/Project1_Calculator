```python
import pandas as pd
import numpy as np
from scipy.stats import boxcox
from sklearn.preprocessing import PowerTransformer

# Example: Find numeric columns
numeric_cols = merged_df.select_dtypes(include=[np.number]).columns

# Check skewness
skewness = merged_df[numeric_cols].skew()
print(skewness)

# Apply transformation to reduce skewness:
# Option 1: Log transformation (for positive values)
for col in numeric_cols:
    if (merged_df[col] > 0).all():
        merged_df[col + '_log'] = np.log1p(merged_df[col])

# Option 2: Box-Cox transformation (for positive values)
for col in numeric_cols:
    if (merged_df[col] > 0).all():
        merged_df[col + '_boxcox'], _ = boxcox(merged_df[col])

# Option 3: Yeo-Johnson transformation (works with positive and negative values)
pt = PowerTransformer(method='yeo-johnson')
merged_df[numeric_cols] = pt.fit_transform(merged_df[numeric_cols])

# Check skewness after transformation
print(merged_df[numeric_cols].skew())
```
